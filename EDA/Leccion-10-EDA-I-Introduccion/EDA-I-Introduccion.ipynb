Donde variable es cada una de las variables categóricas que tenemos en el dataset y n_unicos es una lista con los nombres de los valores únicos de cada variable.

Valores únicos y su frecuencia

```python
#frecuencia
for col in df_categoricas.columns:
    print("--------------")
    print(f"En la columna '{col}' tenemos los siguientes valores:")
    print(pd.DataFrame(df[col].value_counts()))
    print("--------------")
```

El output de las frecuencias es un dataframe, pero printeado en forma de serie en el cual vemos por cada valor único el número de veces que se repite.

Bonus

```python
valores_unicos = {col:df[col].unique().tolist() for col in df_categoricas.columns}
```

```python
df_2 = pd.DataFrame(columns=["variable", "n_uniques"])
df_2["variable"] = valores_unicos.keys()
df_2["n_uniques"] = valores_unicos.values()
```
