### Objetivos
Usad la librería sidetable para:

1. Explorar los valores nulos.

```python
df.stb.missing()
```

Explorando los nulos de nuestro dataframe vemos que:

.- mes = Nos encontramos un 12.83% de nulos.

.- age = Nos encontramos un 8.99% de nulos.

.- species_ = Nos encontraoms un 7.3% de nulos.

.- fatal = Nos encontramos un 5.8% de nulos.

.- activity = Nos encontramos un 1.8% de nulos.

.- sex, country = No llegan al 1% de nulos.

.- year = No tiene nulos.

En conclusión, visualizamos que no hay un porcentaje elevado de nulos en nuestro dataframe.

2. Explorad la columna de country y reportad los resultados solo para aquellas categorías que incluyan el 70% de los datos.

```python
df.stb.freq(["country"])
```

En esta visualización vemos que el 70% de nuestros datos en la columna "country" se agrupan en solo dos países: usa y australia.

```python
df.stb.freq(["country"], thresh= 70, other_label="others countrys")
```

Con el parámetro thresh podemos visualizar mejor los dos países que se encuentran en el 70% de nuestros datos. Los demás países los agrupamos en la etiqueta "others countrys"

3. Explorad la columna country en relación al sexo, filtrando de nuevo por aquellas categorías que incluyan el 70% de los datos.

```python
df.stb.freq(["country", "sex"], thresh= 70, other_label="others countrys")
```

Visualizando nuestro sidetable vemos que solo hay tres países que se agrupen en el 70% de los datos (usa, australia y south africa) y solo tienen registro de ataques a hombres.

4. Haced los mismo para la columna activity en relación a la columna fatal. En este caso filtrad y mostrad los datos que incluyan el 60% de los datos.

```python
df.stb.freq(["activity", "fatal"], thresh= 60, other_label="others activities")
```

Visalizando nuestro dataframe con sidetable comprobamos que nos aparecen muchas actividades pero en relación a la columna "Fatal", las actividades que tienen accidentes mortales son "swimming" y "surfing", mientras que las otras sufres ataques de tiburones pero con poca probabilidad de mortalidad.

En el caso de ataques vemos que el surfing tiene un 24.51% de ataques no mortales de tiburones frente a un 1.68% de ataques mortales de tiburones. En cuanto al swimming vemos que tiene un 8.99% de ataques no mortales de tiburones frente a un 5.04% de ataques mortales de tiburones. 

Sacamos en conclusión que hay pocos ataques mortales realizando las diferentes actividades.

5. Por último, realizad este análisis para la columna fecha (la que creamos en el último ejercicio de pair programming de Pandas) en relación a la columna fatal. Mostrad solo los resultados para las categorías que incluyan el 60% de los datos.

```python
df.stb.freq(["mes", "fatal"], thresh= 60)
```

Visualizando este sidetable en relación a la columna "mes" y "fatal" vemos que en los meses más cálidos comprendidos entre Abril y Octubre es cuando se dan más ataques de tiburones.
