### Objetivos
Usad la librer√≠a sidetable para:

1. Explorar los valores nulos.

```python
df.stb.missing()
```

Explorando los nulos de nuestro dataframe vemos que:

.- mes = Nos encontramos un 12.83% de nulos.

.- age = Nos encontramos un 8.99% de nulos.

.- species_ = Nos encontraoms un 7.3% de nulos.

2. Explorad la columna de country y reportad los resultados solo para aquellas categor√≠as que incluyan el 70% de los datos.

```python
df.stb.freq(["country"])
```

En esta visualizaci√≥n vemos que el 70% de nuestros datos en la columna "country" se agrupan en solo dos pa√≠ses: usa y australia.

```python
df.stb.freq(["country"], thresh= 70, other_label="others countrys")
```

Con el par√°metro thresh podemos visualizar mejor los dos pa√≠ses que se encuentran en el 70% de nuestros datos. Los dem√°s pa√≠ses los agrupamos en la etiqueta "others countrys"

3. Explorad la columna country en relaci√≥n al sexo, filtrando de nuevo por aquellas categor√≠as que incluyan el 70% de los datos.

```python
df.stb.freq(["country", "sex"], thresh= 70, other_label="others countrys")
```

Visualizando nuestro sidetable vemos que solo hay tres pa√≠ses que se agrupen en el 70% de los datos (usa, australia y south africa) y solo tienen registro de ataques a hombres.

4. Haced los mismo para la columna activity en relaci√≥n a la columna fatal. En este caso filtrad y mostrad los datos que incluyan el 60% de los datos.

```python
df.stb.freq(["activity", "fatal"], thresh= 60, other_label="others activities")
```

Visalizando nuestro dataframe con sidetable comprobamos que nos aparecen muchas actividades pero en relaci√≥n a la columna "Fatal", las actividades que tienen accidentes mortales son "swimming" y "surfing", mientras que las otras sufres ataques de tiburones pero con poca probabilidad de mortalidad.
