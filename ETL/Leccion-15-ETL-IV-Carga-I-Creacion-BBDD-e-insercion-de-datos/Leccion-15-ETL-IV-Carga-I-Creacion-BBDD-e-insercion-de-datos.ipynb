```python
# Importamos librer칤a
def conectar_bbdd(password):
    # Funci칩n para comprobar si se establece la conexi칩n con el servidor
    # Par치metro password: string con la contrase침a del servidor
    try:
        cnx = mysql.connector.connect(user = 'root', password = password, host = '127.0.0.1')
    # en el caso de que haya errores
    except mysql.connector.Error as err:
        if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
            print("Something is wrong with your user name or password")
        elif err.errno == errorcode.ER_BAD_DB_ERROR:
            print("Database does not exist")
        else:
            print(err)
    else:
        cnx.close()
        # Si devuelve close todo ha funcionado correctamente
        return "close"
```

Tenemos la conexi칩n correcta con el servidor de SQL

```python
def crear_bbdd(nombre_bbdd, password):
    # Funci칩n para crear una base de datos
    # Par치metro nombre_bbdd: string con el nombre de la base de datos a crear
    # password: string con la contrase침a del servidor
    mydb = mysql.connector.connect(user = 'root', password = f"{password}", host = '127.0.0.1')
    #print("Conexi칩n realizada con 칠xito")
    mycursor = mydb.cursor()
    try:
        mycursor.execute(f"CREATE DATABASE IF NOT EXISTS {nombre_bbdd};")
        print(mycursor)
    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
    else:
        mycursor.close()
        mydb.close()
```

```python
my_bbdd = "tiburones"
```

```python
crear_bbdd(my_bbdd, contrase침a)
```

```python
def crear_insertar_tabla(nombre_bbdd, password, query):
    # Funci칩n para crear una tabla dentro de una base de datos
    # Par치metro nombre_bbdd: string con el nombre de la base de datos a insertar la tabla
    # password: string con la contrase침a del servidor
    # query: consulta en SQL de creaci칩n de tabla
    cnx = mysql.connector.connect(user = 'root', password = password, host = '127.0.0.1', database = nombre_bbdd)
    mycursor = cnx.cursor()
    try:
        mycursor.execute(query)
        print(mycursor)
        cnx.commit()
    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
    else:
        mycursor.close()
        cnx.close()
```

```python
tabla_ataques = """ CREATE TABLE IF NOT EXISTS `tabla_ataques`(
                    `level_0` FLOAT NOT NULL,
                    `type` VARCHAR(50), 
                    `country` VARCHAR(50), 
                    `age` VARCHAR(50),
                    `species_` VARCHAR(50),
                    `fatal_Unknown` VARCHAR(50), 
                    `sex` VARCHAR(50), 
                    `species_.1` VARCHAR(50), 
                    `year` FLOAT, 
                    `age_NORM` FLOAT, 
                    PRIMARY KEY (`level_0`))
                    ENGINE = InnoDB; """
```

```python
tabla_clima = """ CREATE TABLE IF NOT EXISTS `tabla_clima`(
                    `level_0` FLOAT NOT NULL,
                    `timepoint` INT, 
                    `cloudcover` INT, 
                    `highcloud` INT,
                    `midcloud` INT,
                    `lowcloud` INT,
                    `temp2m` INT, 
                    `lifted_index` INT, 
                    `rh2m` VARCHAR(50), 
                    `msl_pressure` INT,
                    `prec_amount` INT,
                    `snow_depth` INT,
                    `wind10m.direction` VARCHAR(50), 
                    `wind10m.speed` INT, 
                    `pais` VARCHAR(50),
                    PRIMARY KEY (`level_0`),
                    CONSTRAINT `fk_ataques_clima`
                    FOREIGN KEY (`level_0`)
                    REFERENCES `tabla_ataques` (`level_0`))
                    ENGINE = InnoDB; """
```

```python
crear_insertar_tabla(my_bbdd, contrase침a, tabla_ataques)
```

```python
crear_insertar_tabla(my_bbdd, contrase침a, tabla_clima)
```
